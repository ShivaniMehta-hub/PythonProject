```python
import math as m
import matplotlib.pyplot as plt
import csv
    
x = []
prices=[]
with open('carprices.txt','r') as textfile:
    plots = csv.reader(textfile)
    for row in plots:
        x.append(row[0])
    x.pop(0)
    for i in x:
        try:
            if i.isdigit() == False:
                raise ValueError
        except ValueError:
             print('This data is invalid and will be ignored',i)
        else:
            prices.append(m.floor(int(i)/10))
                    
plt.hist(prices, bins = 50, histtype = 'bar', rwidth = 0.7) 
plt.xlabel('Prices in tens of thousands')
plt.ylabel('Number of Cars')
plt.title('Frequency of Car Prices')
plt.show()
```

```python

import matplotlib.pyplot as plt
import csv
    
x = []
cartype=[]
with open('cartype.txt','r') as textfile:
    plots = csv.reader(textfile)
    for row in plots:
        x.append(row[0])
    x.pop(0)
cartype = list(set(x))
slices = []
for item in cartype:
    slices.append(x.count(item))
colors = [ 'magenta', 'blue', 'powderblue'] 
plt.pie(slices, labels = cartype, colors=colors, explode = (0.04, 0.04, 0.04, 0.04, 0.04, 0.04), 
        startangle=90, radius = 1.3, autopct = '%2.2f%%') 

plt.show() 
```

```python
import re
import matplotlib.pyplot as plt
import csv
    
Year = []
BearPopulation=[]
WhalePopulation=[]
DolphinPopulation=[]

fin = open("wildlife.txt", "rt")
fout = open("wildlifeout.txt", "wt")
for line in fin:
    fout.write(re.sub('\t+',' ',line))
fin.close()
fout.close()

with open('wildlifeout.txt','rt') as textfile:
    plots = csv.reader(textfile, delimiter=' ')
    for row in plots:
        Year.append(row[0])
        BearPopulation.append(row[1])
        DolphinPopulation.append(row[2])
        WhalePopulation.append(row[3])
        
Year.pop(0)
BearPopulation.pop(0)
DolphinPopulation.pop(0)
WhalePopulation.pop(0)
BearPopulation_list = list(map(int, BearPopulation)) 
DolphinPopulation_list = list(map(int, DolphinPopulation)) 
WhalePopulation_list = list(map(int, WhalePopulation)) 
    
plt.plot(Year, BearPopulation_list, label = "Bears") 
plt.plot(Year, DolphinPopulation_list, label = "Dolphins") 
plt.plot(Year, WhalePopulation_list, label = "Whales") 
plt.xlabel('Years') 
plt.ylabel('Wildlife Population')
plt.title('Wildlife Population Per Year')   
plt.legend(loc='upper left') 
plt.show() 
```

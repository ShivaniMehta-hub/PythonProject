```python
class Employee:

    def __init__(self, name, address, vehicle1 ):
        self.__name = name
        self.__address = address
        self.__emp_vehicle = vehicle1

    def get_name(self):
        return self.__name

    def get_address(self):
        return self.__address
    
    def get_vehicleData(self):
        return self.__emp_vehicle

    def set_name(self, name):
        self.__name = name
    
    def set_address(self, address):
        self.__address = address
        
    def set_vehicleData(self, vehicle1):
        self.__emp_vehicle = vehicle1
        
    def __str__(self):
        emp_str = "Employee Name: {0}; Employee Address: {1}".format(self.__name, self.__address)
        return emp_str
```

```python
class Vehicle:

    def __init__(self, make, model, yom, mileage):
        self.__make = make
        self.__model = model
        self.__yom = yom
        self.__mileage = mileage

    def get_make(self):
        return self.__name

    def get_model(self):
        return self.__model
    
    def get_yom(self):
        return self.__yom
    
    def get_mileage(self):
        return self.__mileage

    def set_make(self, make):
        self.__make = make
    
    def set_model(self, model):
        self.__model = model
        
    def set_yom(self, yom):
        self.__yom = yom
    
    def set_mileage(self, mileage):
        self.__mileage = mileage
        
    def __str__(self):
        vehicle_str = 'Make: '+ self.__make + '; Model: ' + self.__model + '; Year of Manufacturing:' + str(self.__yom) + '; Mileage ' + str(self.__mileage)
        return str(vehicle_str)
```

```python
class FullTimeEmployee(Employee):
    def __init__(self, name, address, vehicleData, salary):
        super().__init__(name, address, vehicleData)
        self.__salary = salary
    
    def get_salary(self):
        return self.__salary
        
    def set_salary(self, salary):
        self.__salary = salary
    
    def __str__(self):
        print('Details of Full time employee:')
        op = super().__str__() #we call the parent class's __str__method first.
        op += "; Salary: {} ".format(self.__salary)
        return op  
   
    #This is the new method added to the Full-Time-Employee class.
    def compute_compensation(self):
        
        if self.__salary <= 45000:
            taxes = (self.__salary * 0.18) 
            compensation=self.__salary - taxes
        elif 45000 < self.__salary <= 82000:
            taxes = (45000 * 0.18) + (self.__salary - 45000) * 0.28
            compensation=self.__salary - taxes
        elif self.__salary > 82000:
            taxes = (45000 * 0.18) + (37000 * 0.28) + (self.__salary - 82000) * 0.33
            compensation = self.__salary - taxes
        return compensation
                
```

```python
class HourlyEmployee(Employee):
    
    def __init__(self, name, address, vehicleData, hoursWorked, hourlyRate):
        super().__init__(name, address, vehicleData)
        self.__hoursWorked = hoursWorked
        self.__hourlyRate = hourlyRate
    
    def get_hoursWorked(self):
        return self.__hoursWorked
    
    def get_hourlyRate(self):
        return self.__hourlyRate
        
    def set_hoursWorked(self, hoursWorked):
        self.__hoursWorked = hoursWorked
        
    def set_hourlyRate(self, hourlyRate):
        self.__hourlyRate = hourlyRate
    
    def __str__(self):
        print('Details of Hourly employee:')
        op = super().__str__() #we call the parent class's __str__method first.
        op += "; Hours worked: {} ".format(self.__hoursWorked)
        op += "; Hourly Rate: {} ".format(self.__hourlyRate)
        return op    
   
    #This is the new method added to the Hourly Employee class.
    def compute_compensation(self):
        if self.__hoursWorked <= 40:
            compensation = self.__hoursWorked * self.__hourlyRate
        else:
            compensation = (40 * self.__hourlyRate) + (self.__hoursWorked - 40) * self.__hourlyRate * 1.8
        return compensation
```

```python
class Consultant(Employee):
    
    def __init__(self, name, address, vehicleData, hoursWorked, projectType):
        super().__init__(name, address, vehicleData)
        self.__hoursWorked = hoursWorked
        self.__projectType = projectType
 
    def get_hoursWorked(self):
        return self.__hoursWorked
    
    def get_projectType(self):
        return self.__projectType
        
    def set_hoursWorked(self, hoursWorked):
        self.__hoursWorked = hoursWorked
        
    def set_projectType(self, projectType):
        self.__projectType = projectType
    
    def __str__(self):
        print('Details of Consultant:')
        op = super().__str__() #we call the parent class's __str__method first.
        op += "; Hours Worked: {} ".format(self.__hoursWorked)
        op += "; Project Type: {} ".format(self.__projectType)
        return op    
   
    #This is the new method added to the Consultant class.
    def compute_compensation(self):
        if self.__projectType == 1:
             hourlyRate = 55.00
        elif self.__projectType == 2:
             hourlyRate = 70.00
        elif self.__projectType == 3:
             hourlyRate = 85.00
        
        compensation = self.__hoursWorked * hourlyRate
        return compensation
        
```

```python
def add_employee():
    try:
            emp_type = int(input('Type of Employee to be added ? (1 = Full-Time, 2 = Hourly, 3 = Consultant)'))
            if emp_type==1:
                name,address = emp_detail()
                make,model,yom,mileage = vehicle_data()
                vehicleData = Vehicle(make,model,yom,mileage)
                try:
                    salary = int(input('Enter the employee salary:'))
                except:
                    print('Enter the integer value for salary !')
                    salary = int(input('Enter the employee salary:'))
                
                emp_obj = FullTimeEmployee(name, address, vehicleData, salary) 
            
            elif emp_type==2:
                name,address = emp_detail()
                make,model,yom,mileage = vehicle_data()
                vehicleData = Vehicle(make,model,yom,mileage)
                try:
                    hoursWorked = int(input('Enter the hours worked:'))
                except:
                    print('Enter the integer value for hours Worked !')
                    hoursWorked = int(input('Enter the hours worked:'))
                try:
                    hourlyRate = int(input('Enter the hourly rate:'))
                except:
                    print('Enter the integer value for hourly Rate !')
                    hourlyRate = int(input('Enter the hourly rate:'))
                
                emp_obj = HourlyEmployee(name, address, vehicleData, hoursWorked, hourlyRate)
            
            elif emp_type==3:
                name,address = emp_detail()
                make,model,yom,mileage = vehicle_data()
                vehicleData = Vehicle(make,model,yom,mileage)
                try:
                    hoursWorked = int(input('Enter the hours worked:'))
                except:
                    print('Enter the integer value for hours Worked !')
                    hoursWorked = int(input('Enter the hours worked:'))
                try:
                    projectType = int(input('Enter the project type:'))
                    if projectType == 1 or projectType == 2 or projectType == 3:
                        pass
                except:
                    print('Enter the integer value of 1,2 or 3 for project type !')
                    projectType = int(input('Enter the project Type:'))
                
                emp_obj = Consultant(name, address, vehicleData, hoursWorked, projectType)
                
            if os.path.exists('empdata.dat'):
                with open('empdata.dat', 'ab') as employee_file:
                    pickle.dump(emp_obj, employee_file)
            else:
                with open('empdata.dat', 'wb') as employee_file:
                    pickle.dump(emp_obj, employee_file)
    except:
        print('Employee type can only be 1, 2 or 3')
        add_employee()
            
```

```python
def emp_detail():
    name = input('Enter the employee name:')
    address = input('Enter the employee address:')
    return (name,address)
                            
def vehicle_data():
    make = input('Enter the employee vehicle make:')
    model = input('Enter the employee vehicle model:')
    try:
        yom = int(input('Enter the vehicle year of manufacture:'))
    except:
        print('Enter the integer value for year !')
        yom = int(input('Enter the vehicle year of manufacture:'))
    try:
        mileage = int(input('Enter the vehicle mileage:'))
    except:
        print('Enter the integer value for mileage !')
        mileage = int(input('Enter the vehicle mileage:'))
    return (make,model,yom,mileage)
```

```python
def read_emp_file():
    my_list = [] #Create an empty list
    try:
        emp = open('empdata.dat', 'rb') 
    except FileNotFoundError:
        return my_list 
    else:  
        try:
            while (True): 
                 my_list.append(pickle.load(emp))
        except EOFError: 
            emp.close()
            return my_list
                
def print_emp_compensation(my_list):
    for obj in my_list:
        print('{0} \t\t${1:0.2f}'.format(obj.get_name(), obj.compute_compensation()))

def print_emp_vehicle_info(my_list):
    for obj in my_list:
        if obj.get_vehicleData().get_mileage() > 78000:
            print('Employee name is:''{0}   {1}'.format(obj.get_name(), obj.get_vehicleData()))

def exit_emp(my_list): 
    print('You chose to exit the program')
    exit_input = input('Are you sure (Y/N)?')
    if exit_input == 'Y' or exit_input == 'y':
            ed = open('empdata.dat', 'wb')
            for obj in my_list:
                pickle.dump(obj, ed)
            ed.close()
            sys.exit()
    else:
            menu()
```

```python
import os.path
import sys
import pickle

def menu():
    emp_list = read_emp_file()
    print("=======MAIN MENU=======")
    print()
    choice = input("""
                    1. Add a new employee 
                    2. Print all employee data for all the employees in the file 
                    3. Compute and print the employee name and compensation of all employees 
                    4. Print the employee name with vehicle information vehicles with mileage greater than 78000 miles 
                    5. Exit the application

                      Your selection: """)
    
    if choice == '1':
        add_employee()
        menu()
        
    elif choice == '2':
        if len(emp_list) == 0: 
            print("No employees in database!")
            menu()
        else:
            for obj in emp_list:
                print(obj.__str__())
                print(obj.get_vehicleData())
                print(' ')
                
    elif choice == '3':
        if len(emp_list) == 0: 
            print("No employees in database!")
            menu()
        else:
            print("\nEmployeeName\tCompensation")
            print("========================================")
            print_emp_compensation(emp_list)
            
    elif choice == '4':
        if len(emp_list) == 0: 
            print("No employees in database!")
            menu()
        else:
            print("\nEmployee Name and Vehicle Data for high mileage vehicles")
            print("=========================================")
            print_emp_vehicle_info(emp_list)
            
    elif choice == '5':
        exit_emp(emp_list)
    else:
        print("You must enter an integer between 1 and 5 !")
        print("Please try again")
        menu()
        
def main():
    menu()  
    
main()
```
